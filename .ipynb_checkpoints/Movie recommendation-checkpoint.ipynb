# Creating our model
Now we will create our movie recommender

```python
movie_matrix = df3.pivot_table(index="user_id", columns="title", values="rating")
```

```python
movie_matrix
```

```python
ratings.sort_values('num of ratings', ascending = False).head()
```

```python
starwars_rating = movie_matrix['Star Wars (1977)']
starwars_rating.head()
```

```python
similar_to_starw = movie_matrix.corrwith(starwars_rating)
```

```python
correlation_with_starw = pd.DataFrame(similar_to_starw, columns =['Correlation'])
```

```python
correlation_with_starw.dropna(inplace = True)
```

```python
correlation_with_starw.head()
```

```python
correlation_with_starw.sort_values('Correlation', ascending = False).head(10)
```

## Filteration of correlated data
We only want to correlate with those movies which have been watched/rated by atleast 100 users.

```python
corr_StarWars = correlation_with_starw.join(ratings['num of ratings'])
```

```python
corr_StarWars.head()
```

```python
corr_StarWars[corr_StarWars['num of ratings'] > 100].sort_values('Correlation', ascending=False)
```

```python
def predict_movies(movie_name):
    movie_user_rating = movie_matrix[movie_name]
    similar_to_movie = movie_matrix.corrwith(movie_user_rating)
    
    correlation_movie = pd.DataFrame(similar_to_movie, columns =['Correlation'])
    correlation_movie.dropna(inplace = True)
    
    correlation_movie = correlation_movie.join(ratings['num of ratings'])
    
    prediction = correlation_movie[correlation_movie['num of ratings'] > 100].sort_values('Correlation', ascending=False)
    
    return prediction
    
```

```python
prediction = predict_movies("Titanic (1997)")
```
